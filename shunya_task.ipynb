#Image 1

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

## Reading the image
img = cv2.imread("/content/gdrive/MyDrive/shunya1.png")

## Plot the image using matplotlib for pixel examination
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
print(img.shape)
```

```python
## Assigning the coordinates of the object
pts = np.array([[110,50],[530,80],[600,740],[50,780]])

## (1) Crop the bounding rect
rect = cv2.boundingRect(pts)
x,y,w,h = rect
croped = img[y:y+h, x:x+w].copy()

## (2) make mask
pts = pts - pts.min(axis=0)

mask = np.zeros(croped.shape[:2], np.uint8)
cv2.drawContours(mask, [pts], -1, (255, 255, 255), -1, cv2.LINE_AA)

## (3) do bit-op
dst = cv2.bitwise_and(croped, croped, mask=mask)

## (4) add the white background
bg = np.ones_like(croped, np.uint8)*255
cv2.bitwise_not(bg,bg, mask=mask)
dst2 = bg+ dst

## (5) output the images
cv2_imshow(croped)
cv2_imshow(mask)
cv2_imshow(dst)
cv2_imshow(dst2)
```

The required output

```python
pts = np.array([[110,50],[530,80],[600,740],[50,780]])
## Creating a polygon around the object
output_img = cv2.polylines(img, [pts], True, [255, 0, 0], 1)
print('OUTPUT IMAGE 1')
cv2_imshow(output_img)
```
